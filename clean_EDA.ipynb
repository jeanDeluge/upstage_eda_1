### 침체 시작 1년전과 장단기 금리차이 

```python
# 0에서 1로 바뀌는 순간 : 침체 시작일
transitions = data[(data['USA_recession'].shift(1) == 0) & (data['USA_recession'] == 1)]

# 침체 시작일 9개월 전 부터의 10년물-2년물 금리차이
transition_periods = []
for date in transitions.index:
    start_date = date - pd.DateOffset(days=365)
    period_data = data[(data.index >= start_date) & (data.index < date)]
    transition_periods.append(period_data)

# Displaying the number of transition periods and the first transition period data
print(transitions)
```

```python
# Plotting the '10y-2y' values for each transition period
plt.figure(figsize=(20, 10), facecolor='white')
for i, period in enumerate(transition_periods, 1):
    plt.plot(period.index, period['10y-2y'], label=f'1y before recession {i}')

# Plotting the '10y-2y' values for non-transition periods
non_transition_data = data[~data.index.isin([p.index[0] for p in transition_periods])] 
plt.plot(non_transition_data.index, non_transition_data['10y-2y'], label='Non-Transition', color='grey', alpha=0.5)

plt.title('10y-2y 1y Before Transition vs. Non-Transition Periods')
plt.xlabel('Date')
plt.ylabel('10y-2y')
plt.legend()
plt.axhline(y=0, color='red')
plt.show()
```

```python
# Plotting the 'NASDAQ' values for each transition period
plt.figure(figsize=(20, 10), facecolor='white')
for i, period in enumerate(transition_periods, 1):
    plt.plot(period.index, period['NASDAQ'], label=f'1y before recession {i}')

# Plotting the 'NASDAQ' values for non-transition periods
non_transition_data = data[~data.index.isin([p.index[0] for p in transition_periods])] 
plt.plot(non_transition_data.index, non_transition_data['NASDAQ'], label='Non-Transition', color='grey', alpha=0.5)

plt.title('NASDAQ 1y Before Transition vs. Non-Transition Periods')
plt.xlabel('Date')
plt.ylabel('NASDAQ')
plt.legend()
plt.axhline(y=0, color='red')
plt.show()
```

```python
# Plotting the 'NASDAQ 등락률' values for each transition period
plt.figure(figsize=(20, 10), facecolor='white')
for i, period in enumerate(transition_periods, 1):
    plt.plot(period.index, period['NASDAQ_fluc'], label=f'1y before recession {i}')

# Plotting the 'NASDAQ 등락률' values for non-transition periods
non_transition_data = data[~data.index.isin([p.index[0] for p in transition_periods])] 
plt.plot(non_transition_data.index, non_transition_data['NASDAQ_fluc'], label='Non-Transition', color='grey', alpha=0.5)

plt.title('NASDAQ fluctuation 1y Before Transition vs. Non-Transition Periods')
plt.xlabel('Date')
plt.ylabel('NASDAQ fluctuation')
plt.legend()
plt.axhline(y=0, color='red')
plt.show()
```
